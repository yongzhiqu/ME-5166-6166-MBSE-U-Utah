```python
e = 2.7182818284590452353602874713526624977572
def cal_sin_x(x):
    sin_x=(e**(x*1j)).imag
    return(sin_x)
```

```python
n = 0
A = 500000000000.0
m = -4.0
def increase_n(n):
    n = n+1
    return(n)
n=increase_n(n)
print(n)
```

```python
def cal_sigma_a(n):
    sigma_a = 120 + 20 * cal_sin_x(n)
    return(sigma_a)

def cal_nf(sigma_a, A, m):
    nf = A*sigma_a**(m)
    return(nf)

def cal_D(nf):
    return(1.0/nf)

sigma = cal_sigma_a(n)
nf = cal_nf(sigma, A, m)
D = cal_D(nf)
percentage = 1-D


while percentage > 0.1:
      n = increase_n(n)
      percentage = percentage - D
      print(n)
      print(percentage)
```
